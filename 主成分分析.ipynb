```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.datasets import load_boston
sns.set()
```

```python
boston = load_boston()
df = boston.data
df = pd.DataFrame(df, columns=boston.feature_names)
df
```

```python
# 主成分分析実装
pca = PCA()
pca_data = pca.fit(df)
pca_data = pca.transform(df)
```

```python
pd.DataFrame(pca_data)
```

```python
plt.figure(figsize=(10,6))
plt.hist(pca_data[:,0])
plt.xlabel("PCA1")
plt.show()
```

```python
pd.DataFrame(pca.explained_variance_ratio_ * 100)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pca_data[:,0], pca_data[:,1])
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

```python
df_pca = pd.concat([pd.DataFrame(df[pca_data[:,0] > 100].mean()), pd.DataFrame(df[pca_data[:,0] <= 100].mean())], axis=1)
# カラム名を修正
df_pca.columns = ["第1主成分が100より大きい", "第1主成分が100以下"]
df_pca
```

```python
df_pca = pd.concat([pd.DataFrame(df[pca_data[:,1] > 100].mean()), pd.DataFrame(df[pca_data[:,1] <= 100].mean())], axis=1)
# カラム名を修正
df_pca.columns = ["第2主成分が100より大きい", "第2主成分が100以下"]
df_pca
```
